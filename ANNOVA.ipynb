# 1. For the suburb Altona, it is postulated that a typical property sells for 800,000. Use the data at hand to test this assumption. Is the typical property price really 800,000 or has it increased? Use a significance level of 5%.

```python
altona_prices = df[df['Suburb']=='Altona']['Price']

#assumed population mean

population_mean= 800000

#Set signifance value
alpha= 0.05

# Perform t-test
t_statistic, p_value = stats.ttest_1samp(altona_prices, population_mean)

# Calculate degrees of freedom
degrees_of_freedom = len(altona_prices) - 1

# Find the critical t-value
critical_t_value = stats.t.ppf(1 - alpha/2, degrees_of_freedom)

# Make a decision
if abs(t_statistic) > critical_t_value:
    print("Reject the null hypothesis: The typical property price has changed.")
else:
    print("Fail to reject the null hypothesis: The typical property price is $800,000.")

# Print the t-statistic, p-value, and critical t-value
print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")
print(f"Critical t-value: {critical_t_value}")
```

# 2. For the year 2016, is there any difference in prices of properties sold in the summer months vs winter months? Consider months from October till March as winter months and rest as summer months. Use a significance level of 5%. 

```python
# Convert the "Date" column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

# Extract the year and month from the "Date" column
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# Define summer and winter months (October to March as winter)
winter_months = [10, 11, 12, 1, 2, 3]
summer_months = [4, 5, 6, 7, 8, 9]

# Filter the data for the year 2016
data_2016 = df[df['Year'] == 2016]

# Extract property prices for summer and winter months in 2016
winter_prices = data_2016[data_2016['Month'].isin(winter_months)]['Price']
summer_prices = data_2016[data_2016['Month'].isin(summer_months)]['Price']

# Set the significance level
alpha = 0.05

# Perform a two-sample t-test
t_statistic, p_value = stats.ttest_ind(winter_prices, summer_prices, equal_var=False)

# Make a decision 
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in property prices between summer and winter months in 2016.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in property prices between summer and winter months in 2016.")

# Print the t-statistic and p-value
print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")
```

# 3. For the suburb Abbotsford, what is the probability that out of 10 properties sold, 3 will not have a car parking? Use the column car in the dataset. Round off your answer to 3 decimal places.

```python
 
# Filter properties in the suburb Abbotsford
abbotsford_properties = df[df['Suburb'] == 'Abbotsford']

# Calculate the probability of having no car parking in Abbotsford
p_no_car_parking = (abbotsford_properties['Car'] == 0).mean()

# Define the values for the binomial distribution
n = 10  # Total number of properties sold
k = 3   # Number of properties without car parking

# Calculate the probability using the binomial distribution formula
probability = stats.binom.pmf(k, n, p_no_car_parking)

# Round the probability to 3 decimal places
probability = round(probability, 3)

print(f"The probability that out of 10 properties sold in Abbotsford, 3 will not have car parking is {probability}")
```

# 4. In the suburb Abbotsford, what are the chances of finding a property with 3 rooms? Round your answer to 3 decimal places.

```python

# Filter properties in the suburb Abbotsford with 3 rooms
abbotsford_3_rooms = df[(df['Suburb'] == 'Abbotsford') & (df['Rooms'] == 3)]

# Calculate the probability of finding a property with 3 rooms in Abbotsford
probability = len(abbotsford_3_rooms) / len(df[df['Suburb'] == 'Abbotsford'])

# Round the probability to 3 decimal places
probability = round(probability, 3)

print(f"The chances of finding a property with 3 rooms in Abbotsford is {probability}")
```
